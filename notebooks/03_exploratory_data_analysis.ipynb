```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_customer_data.csv")

df.head()
```

```python
df.describe()
```

```python
total_revenue = df["purchase_amount_usd"].sum()
print("Total Revenue:", total_revenue)
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby("category")["purchase_amount_usd"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../output/charts/revenue_by_category.png")
plt.show()
```

### Insight: Revenue by Category


- Clothing category generates the highest revenue among all categories  
- Indicates strong and consistent customer demand in this segment  
- Plays a major role in overall sales performance  
- Opportunity to increase revenue through product expansion and targeted promotions 

```python
location_sales = df.groupby("location")["purchase_amount_usd"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
location_sales.plot(kind="bar")
plt.title("Top 10 Locations by Revenue")
plt.xlabel("Location")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../output/charts/top_10_locations_by_revenue.png")
plt.show()
```

### Insight: Revenue by Location

- Certain locations contribute significantly higher revenue compared to others  
- Highlights regional differences in customer purchasing behavior  
- High-performing regions can be targeted for focused marketing and inventory optimization  
- Lower-performing locations may require strategic improvements to boost sales  

```python
gender_sales = df.groupby("gender")["purchase_amount_usd"].sum()

plt.figure(figsize=(6,4))
gender_sales.plot(kind="bar")
plt.title("Revenue by Gender")
plt.xlabel("Gender")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../output/charts/revenue_by_gender.png")
plt.show()
```

### Insight: Revenue by Gender

- Noticeable difference in revenue contribution between genders  
- Reflects variation in purchasing patterns and product preferences  
- Provides opportunity for personalized marketing strategies  
- Helps in designing targeted campaigns for different customer segments

### Insight: Customer Loyalty

- Customers with higher previous purchases show strong repeat buying behavior

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="review_rating", y="purchase_amount_usd", data=df)
plt.title("Rating vs Purchase Amount")
plt.xlabel("Review Rating")
plt.ylabel("Purchase Amount (USD)")
plt.tight_layout()
plt.savefig("../output/charts/rating_vs_purchase_amount.png")
plt.show()
```

### Insight: Ratings vs Purchase Amount

- No strong correlation observed between product ratings and purchase amount  
- Suggests customers are not solely influenced by ratings when purchasing  
- Other factors such as price, need, or brand may drive buying decisions  
- Indicates potential to improve review systems or product positioning  

```python
season_sales = df.groupby("season")["purchase_amount_usd"].sum()

plt.figure(figsize=(6,4))
season_sales.plot(kind="bar")
plt.title("Sales by Season")
plt.xlabel("Season")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../output/charts/sales_by_season.png")
plt.show()
```

### Insight: Seasonal Trends

- Sales vary across seasons, indicating clear seasonal purchasing patterns  
- Certain seasons show higher demand compared to others  
- Helps in identifying peak sales periods  
- Enables better planning for inventory, marketing, and supply chain operations  

### Overall Business Insight

- Clothing is the primary driver of revenue across all categories  
- Customer location and seasonal trends significantly influence sales performance  
- Repeat customers contribute consistently, highlighting importance of retention  
- Data-driven strategies can improve marketing effectiveness and revenue growth
