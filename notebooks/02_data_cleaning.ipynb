```python
import pandas as pd

df = pd.read_csv("../Data/raw/customer_shopping_behavior.csv")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns
```

```python
df.columns = df.columns.str.replace("(", "").str.replace(")", "")
```

```python
categorical_cols = [
    'gender', 'item_purchased', 'category', 'location',
    'size', 'color', 'season', 'subscription_status',
    'shipping_type', 'discount_applied', 'promo_code_used',
    'payment_method', 'frequency_of_purchases'
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df.info()
```

```python
df.to_csv("../data/processed/cleaned_customer_data.csv", index=False)
```
